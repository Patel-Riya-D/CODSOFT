# Scatter plot of sales of TV

```python
x = sales[['TV']]
y = sales['Sales']
```

```python
plt.xlabel('TV')
plt.ylabel('Sales of tv')
plt.scatter(x,y)
```

# Model Train And Test

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=32)
```

```python
print(xtrain)
print(ytrain)
print(xtest)
print(ytest)
```

# Convert data types float to integer

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ytest.values
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(ytest,ypred)
r2
```

**Value of r2 score is 0.81**

```python
plt.figure(figsize=(5,4))
plt.xlabel('TV')
plt.ylabel('Sales of TV')
plt.title('Linear Regression Model')
plt.plot(xtrain,model.predict(xtrain),color='red',label='line of regression')
plt.scatter(xtrain,model.predict(xtrain),color='green',label='Actual Data')
plt.scatter(xtrain,ytrain,label='predicted data')
plt.legend()
plt.show()
```
