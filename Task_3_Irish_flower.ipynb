# Repalce Categorical data to Numerical data

```python
flower.replace({'species':{'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}},inplace=True)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
flower['species']=le.fit_transform(flower['species'])
flower.head()
```

# Train and Test Model

```python
from sklearn.model_selection import train_test_split
x = flower.drop(columns=['species'])
y = flower['species']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

# Test Data Accuracy

```python
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
print("Accuracy: ",model.score(x_test,y_test)*100)
```
